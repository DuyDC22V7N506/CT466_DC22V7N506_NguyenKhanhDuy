```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from preprocessing import load_and_process

# Tải lại dữ liệu và kết quả đã lưu từ main.py
# Giả sử bạn đã chạy main.py và có file output_results.csv
X, df_original = load_and_process("data.xlsx")
df_results = pd.read_csv("output_results.csv")

# Cấu hình chung cho biểu đồ
def plot_cluster(X, labels, title):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='viridis', legend='full')
    plt.title(f'Phân khúc khách hàng - {title}', fontsize=14)
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.legend(title='Cluster')
    plt.show()
```

```python
plot_cluster(X, df_results['KMeans'], "K-Means Clustering")
```

```python
plot_cluster(X, df_results['KMeans++'], "K-Means++ Clustering")
```

```python
plot_cluster(X, df_results['DBSCAN'], "DBSCAN (Density-Based)")
```

```python
plot_cluster(X, df_results['Hierarchical'], "Hierarchical Clustering")
```

```python
plot_cluster(X, df_results['GMM'], "Gaussian Mixture Model")
```

```python
plot_cluster(X, df_results['AutoEncoder'], "AutoEncoder + K-Means")
```

```python
plot_cluster(X, df_results['MLP'], "Multi-Layer Perceptron (Pseudo-labels)")
```
